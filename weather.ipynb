```python
import pandas as pd
import csv
import codecs
import urllib.request
import urllib.error
import sys
from datetime import datetime, timedelta
```

```python
BaseURL = 'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/'
ApiKey = 'T2QGZ4FXK2SVTGU3XZLHU4URU'  # Replace 'YOUR_API_KEY' with your actual API key
UnitGroup = 'us'  # or 'metric' depending on your preference
Location = 'Eugene,%20OR'  # Location for the weather data

# Set start date to today's date
StartDate = datetime.now().strftime('%Y-%m-%d')

# Set end date to today's date + 7 days
EndDate = (datetime.now() + timedelta(days=7)).strftime('%Y-%m-%d')

ContentType = 'csv'  # Output format (CSV or JSON)
Include = 'days'  # Include sections (e.g., days, hours, current, alerts)

# Construct the weather data query
ApiQuery = f"{BaseURL}{Location}/{StartDate}/{EndDate}?unitGroup={UnitGroup}&contentType={ContentType}&include={Include}&key={ApiKey}"
```

```python
ApiQuery = BaseURL + Location
if len(StartDate):
    ApiQuery += "/" + StartDate
    if len(EndDate):
        ApiQuery += "/" + EndDate

ApiQuery += "?"
if len(UnitGroup):
    ApiQuery += "&unitGroup=" + UnitGroup

if len(ContentType):
    ApiQuery += "&contentType=" + ContentType

if len(Include):
    ApiQuery += "&include=" + Include

ApiQuery += "&key=" + ApiKey
```

```python
try:
    CSVBytes = urllib.request.urlopen(ApiQuery)
except urllib.error.HTTPError as e:
    ErrorInfo = e.read().decode()
    print('Error code: ', e.code, ErrorInfo)
    sys.exit()
except urllib.error.URLError as e:
    ErrorInfo = e.read().decode()
    print('Error code: ', e.code, ErrorInfo)
    sys.exit()
```

```python
# Parse the results as CSV
CSVText = csv.reader(codecs.iterdecode(CSVBytes, 'utf-8'))

# Open a CSV file for writing
with open('raw_data/upcoming_weather.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)

    # Write headers
    for RowIndex, Row in enumerate(CSVText):
        if RowIndex == 0:
            writer.writerow(Row)
            FirstRow = Row
        else:
            # Write data rows
            writer.writerow(Row)
            print('Weather in', Row[0], 'on', Row[1])
            for ColIndex, Col in enumerate(Row):
                if ColIndex >= 4:
                    print('   ', FirstRow[ColIndex], '=', Row[ColIndex])
```
